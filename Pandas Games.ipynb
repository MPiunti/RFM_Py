```python
import pandas as pd
import numpy as np
 
rows = np.arange(3,9)
cols = list('ABCDF')
 
df = pd.DataFrame(np.random.randn(6,5),index=rows,columns=cols)
df
```

```python
df.describe(percentiles=[0.25,0.75])
```

### Medodi di accesso al dataframe (loc e iloc)

```python

column_labelB = df['B']

#se volete invece fare riferimento a specifiche righe della tabella l'istruzione diventa:
#con il metodo loc sono utilizzate le etichette delle righe;
rows_label46 = df.loc[4:6]   #da 4 a 6

#se intendete estrarre righe a seconda della loro posizione nella tabella, l'istruzione diventa:
rows_index13 = df.iloc[1:3]
```

```python
print(column_labelB)
```

```python
rows_label46
```

```python
rows_label4_ = df.loc[4:]   #da indice 4 fino alla fine
rows_label4_
```

```python
rows_index13
```

### Modifiche al DF

```python
# modifichiamo la cella alla etichetta riga '6' colonna 'B'
df.loc[6]['B'] = 777.7
 
# modifichiamo la cella alla posizione riga '5'  colonne da 'A' a 'F' (ovvero tutta l' ultima riga)
df.iloc[5]['A':'F'] = 777.7
 
# modifico tutta l'ultima colonna
df['F'] =  777.7
df
```

### MERGE (JOIN)

Merge by attribute and features


prepariamo altri due dataframe di esempio df_a e df_b

```python
raw_data = {
        'subject_id': ['10', '20', '30', '40', '50'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches'],
        'city': ['Rome', 'Milan', 'Beijing', 'London', 'New York']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name', 'city'])
df_a
```

```python
raw_data = {
        'subject_id': ['40', '50', '60', '70', '80'],
        'first_name': ['Alice', 'Ayoung', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Aoni', 'Atiches', 'Balwner', 'Brice', 'Btisan'],
        'country': ['IT', 'IT', 'CP', 'GB', 'USA']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name', 'country'])
df_b
```

```python
pd.merge(df_a, df_b, on='subject_id')
```

```python
pd.merge(df_a, df_b, left_on='subject_id', right_on='subject_id')
```

```python
# inner join
pd.merge(df_a, df_b, on='subject_id', how='inner')
```

```python
# outer
pd.merge(df_a, df_b, on='subject_id', how='outer')
```

```python
# left with suffixes
pd.merge(df_a, df_b, on='subject_id', how='left', suffixes=('_left', '_right'))
```
